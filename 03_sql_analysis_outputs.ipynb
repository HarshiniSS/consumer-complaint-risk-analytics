```python
from pathlib import Path

sql_dir = Path("../sql")
sql_dir.mkdir(exist_ok=True)

sql_file = sql_dir / "complaint_analysis_queries.sql"
```

```python
sql_content = """
-- Query 1: Complaints by Product
SELECT product, COUNT(*) AS total_complaints
FROM consumer_complaints
GROUP BY product
ORDER BY total_complaints DESC;


-- Query 2: Top 20 Companies by Complaint Volume
SELECT company, COUNT(*) AS total_complaints
FROM consumer_complaints
GROUP BY company
ORDER BY total_complaints DESC
LIMIT 20;


-- Query 3: Complaints by State
SELECT state, COUNT(*) AS total_complaints
FROM consumer_complaints
WHERE state IS NOT NULL
GROUP BY state
ORDER BY total_complaints DESC;


-- Query 4: Timely vs Delayed Responses
SELECT timely_response, COUNT(*) AS total
FROM consumer_complaints
GROUP BY timely_response;


-- Query 5: Monthly Complaint Trend
SELECT year_month, COUNT(*) AS total_complaints
FROM consumer_complaints
GROUP BY year_month
ORDER BY year_month;


-- Query 6: Average Response Delay by Product
SELECT product,
       AVG(response_delay_days) AS avg_response_delay
FROM consumer_complaints
WHERE response_delay_days IS NOT NULL
GROUP BY product
ORDER BY avg_response_delay DESC;
"""

with open(sql_file, "w") as file:
    file.write(sql_content)

print("SQL file created successfully:", sql_file)
```

```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine("postgresql+psycopg2://admin:admin123@localhost:5432/complaints_db")
```

```python
query = """
SELECT COUNT(*) AS total_rows
FROM consumer_complaints;
"""

pd.read_sql(query, engine)
```

```python
query = """
SELECT product, COUNT(*) AS total_complaints
FROM consumer_complaints
GROUP BY product
ORDER BY total_complaints DESC;
"""

product_df = pd.read_sql(query, engine)
product_df
```

```python
query = """
SELECT company, COUNT(*) AS total_complaints
FROM consumer_complaints
GROUP BY company
ORDER BY total_complaints DESC
LIMIT 20;
"""

company_df = pd.read_sql(query, engine)
company_df
```

```python
query = """
SELECT state, COUNT(*) AS total_complaints
FROM consumer_complaints
WHERE state IS NOT NULL
GROUP BY state
ORDER BY total_complaints DESC;
"""

state_df = pd.read_sql(query, engine)
state_df.head(10)
```

```python
query = """
SELECT timely_response, COUNT(*) AS total
FROM consumer_complaints
GROUP BY timely_response;
"""

timely_df = pd.read_sql(query, engine)
timely_df
```

```python
query = """
SELECT year_month, COUNT(*) AS total_complaints
FROM consumer_complaints
GROUP BY year_month
ORDER BY year_month;
"""

monthly_df = pd.read_sql(query, engine)
monthly_df.head()
```

```python
query = """
SELECT product,
       AVG(response_delay_days) AS avg_response_delay
FROM consumer_complaints
WHERE response_delay_days IS NOT NULL
GROUP BY product
ORDER BY avg_response_delay DESC;
"""

delay_df = pd.read_sql(query, engine)
delay_df
```

```python
output_dir = Path("../reports/sql_outputs")
output_dir.mkdir(parents=True, exist_ok=True)
```

```python
product_df.to_csv(output_dir / "complaints_by_product.csv", index=False)
company_df.to_csv(output_dir / "top_companies.csv", index=False)
state_df.to_csv(output_dir / "complaints_by_state.csv", index=False)
timely_df.to_csv(output_dir / "timely_response_summary.csv", index=False)
monthly_df.to_csv(output_dir / "monthly_complaint_trend.csv", index=False)
delay_df.to_csv(output_dir / "response_delay_by_product.csv", index=False)

print("SQL output files saved successfully.")
```
