```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
df = pd.read_parquet("../data/processed/complaints_cleaned.parquet")
df.head()
```

```python
engine = create_engine("postgresql+psycopg2://admin:admin123@localhost:5432/complaints_db")
```

```python
df.to_sql(
    "consumer_complaints",
    engine,
    if_exists="replace",
    index=False,
    chunksize=50000,
    method="multi"
)

print("Data loaded into PostgreSQL.")
```

```python
query = """
SELECT product, COUNT(*) AS total_complaints
FROM consumer_complaints
GROUP BY product
ORDER BY total_complaints DESC
LIMIT 10;
"""

pd.read_sql(query, engine)
```
